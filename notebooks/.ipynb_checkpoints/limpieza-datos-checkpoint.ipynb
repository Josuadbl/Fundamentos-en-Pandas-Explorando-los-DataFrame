## 8. Diagnóstico de inconsistencias

### 8.1 Valores únicos en variables categóricas

```python
for col in ['Platform', 'Genre']:
    print(f'\n--- {col} ({df[col].nunique()} únicos) ---')
    print(df[col].value_counts())
```

### 8.2 Rango de años

```python
print(f'Año mínimo: {df["Year"].min()}')
print(f'Año máximo: {df["Year"].max()}')

# Años con pocos registros (posibles errores)
print('\nDistribución de registros por año (valores extremos):')
conteo_years = df['Year'].value_counts().sort_index()
print(conteo_years[conteo_years < 5])
```

### 8.3 Verificación de consistencia en ventas globales

`Global_Sales` debería ser la suma de las ventas regionales.

```python
ventas_calculadas = (df['NA_Sales'] + df['EU_Sales'] + df['JP_Sales'] + df['Other_Sales']).round(2)
diferencia = (ventas_calculadas - df['Global_Sales']).abs()
inconsistentes = diferencia[diferencia > 0.01]
print(f'Registros con Global_Sales inconsistente: {len(inconsistentes)}')
if len(inconsistentes) > 0:
    df.loc[inconsistentes.index, ['Name', 'NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales', 'Global_Sales']].head(10)
```

### 8.4 Valores negativos en ventas

```python
cols_ventas = ['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales', 'Global_Sales']
for col in cols_ventas:
    neg = (df[col] < 0).sum()
    if neg > 0:
        print(f'{col}: {neg} valores negativos')
    else:
        print(f'{col}: sin valores negativos')
```

---
## 10. Informe Interpretativo

### Estructura del dataset
El dataset contiene **16,598 registros** y **11 columnas** con información sobre ventas de videojuegos a nivel mundial. Las variables se distribuyen en:
- **4 columnas categóricas:** `Name`, `Platform`, `Genre`, `Publisher`
- **1 columna entera:** `Rank`
- **6 columnas numéricas float:** `Year`, `NA_Sales`, `EU_Sales`, `JP_Sales`, `Other_Sales`, `Global_Sales`

> **Observación:** La columna `Year` está almacenada como `float64` cuando debería ser un entero o tipo fecha. Esto es consecuencia de los valores faltantes, ya que Pandas convierte columnas enteras a float cuando contienen NaN.

---

### Valores faltantes
Se identificaron **dos columnas con valores nulos:**

| Columna     | Nulos | % del total |
|-------------|------:|------------:|
| `Year`      |   271 |       1.63% |
| `Publisher` |    58 |       0.35% |

**Posibles acciones:**
- `Year`: Imputar con la mediana, con el año más frecuente de esa plataforma, o eliminar los registros si representan un porcentaje aceptable.
- `Publisher`: Marcar como `'Unknown'` o intentar completar manualmente para los títulos más relevantes.

---

### Duplicados
No se encontraron filas completamente duplicadas. Sin embargo, existen registros con el mismo `Name` y `Platform`, que podrían corresponder a relanzamientos o versiones regionales — se debe analizar caso a caso antes de eliminar.

---

### Inconsistencias
- **Rango de años:** El dataset abarca desde 1980 hasta 2020, aunque los años más recientes (2017–2020) tienen muy pocos registros, lo que puede reflejar datos incompletos al momento de la recopilación.
- **Ventas globales:** Se debe verificar si `Global_Sales` coincide con la suma de las ventas regionales (`NA + EU + JP + Other`). Pequeñas diferencias pueden deberse a redondeo.
- **Valores negativos:** No se esperan ventas negativas; cualquier valor por debajo de 0 sería un error de captura.

---

### Conclusión
El dataset presenta una calidad general **aceptable**. El porcentaje total de valores faltantes es inferior al 2%, no hay duplicados exactos y las variables numéricas tienen rangos coherentes. Las principales acciones de limpieza recomendadas son:

1. Tratar los **271 nulos en `Year`** (imputación o eliminación).
2. Tratar los **58 nulos en `Publisher`** (imputación con `'Unknown'`).
3. Convertir `Year` de `float64` a `int` o `datetime` tras resolver los nulos.
4. Revisar los registros con `Name + Platform` duplicados para decidir si conservarlos o eliminarlos.
5. Confirmar la consistencia entre `Global_Sales` y la suma de las ventas regionales.
